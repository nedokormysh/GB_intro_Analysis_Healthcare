* **Исследование рандомизированное. Группы сравнения: 3**

Пациенты в возрасте от 18 до 65 лет, у которых наблюдается выпадение волос, связанное с очаговой алопецией, продолжительностью не менее 6 месяцев и не более 10 лет. Пациенты, которые не получали лечения ранее, должны были иметь выпадение волос как минимум на 50% при скрининге.

Пациенты были рандомизированы в соотношении 1:2:1 (12 мг дважды в день:8 мг дважды в день:плацебо). Рандомизация производилась в зависимости от степени выпадения волос на коже головы в одну из

* **Показания**

Измеряют SALT - показатель потери волос, который определяет количество поверхности кожи головы без волос в определенном квадранте кожи головы, который затем суммируется после применения
множителя, указывающего на вклад площади поверхности кожи головы в конкретный квадрант, для получения общей оценки общей потери волос. Суммарный показатель SALT, равный 0%, означает полный рост волос. Суммарный показатель SALT, равный 100%, означает полное выпадение волос.




* **Стратификации нет**

Хотя "попарная разница между группами лечения и плацебо будет оцениваться с помощью теста Мантеля-Хаензеля (общая разница рисков) с использованием исходного уровня выпадения волос на коже головы (частичное или полное/близкое к полному) **в качестве стратификационного фактора** для каждой группы активного лечения по сравнению с плацебо."

# 3. Найти раздел с конечными точками

 разобрать каждую точку и определить ее тип - первичная/вторичная, клиническая/суррогатная


В течении всего времени исследования было запланировано 10 посещений (10-ое  контроль безопасности для тех, кто откажется от расширенного продолжения лечения). 7 во время лечения.

**Первичной конечной точкой** эффективности является доля пациентов, достигших абсолютной оценки по шкале SALT ≤20 на 24-й неделе (т.е. бинарная конечная точка, обозначающая оценку SALT ≤ 20). Первичная конечная точка определяется путем классификации индивидуальных общих оценок SALT на 24-й неделе (≤ 20 и > 20).

Это клиническая точка.

**Вторичные конечные точки.**

График тестирования вторичных ключевых точек.

secondary.png

1) Первой ключевой вторичной конечной точкой является **процент ответивших по опроснику SPRO на 24-й неделе.** SPRO - это опросник, состоящий из одного пункта, на который отвечает пациент и который предназначен для определения степени
удовлетворенности пациентов с алопецией на момент оценки состояния их волос.

Будут оцениваться следующие вторичные конечные точки ответа SPRO:

* Процент респондентов на 12, 16 и 20 неделях, где респонденты определяются как
"очень довольные" или "довольные"
* Доля пациентов, достигших изменения показателей по шкале SPRO на ≥2 балла
по сравнению с исходным уровнем на 12, 16, 20 и 24 неделях.
Кроме того, непрерывная вторичная конечная точка SPRO - изменение по сравнению с
исходным уровнем на неделях 12, 16, 20 и 24

*Это субъективная оценка, поэтому является суррогатной точкой.*

2) Следующей ключевой вторичной конечной точкой является процент пациентов,
достигших абсолютной оценки SALT ≤20 **Текст, выделенный полужирным шрифтом** (т.е. бинарная конечная точка, обозначающая
оценку SALT ≤ 20) на 20-й, 16-й, 12-й и 8-й неделях.

Будут оцениваться следующие вторичные конечные точки реакции на SALT:

* Доля пациентов, достигших относительного снижения балла SALT не менее
чем на 75% по сравнению с исходным уровнем на неделях 12 и 24
* Доля пациентов, достигших относительного снижения балла по шкале SALT не
менее чем на 90% по сравнению с исходным уровнем на неделях 12 и 24
* Доля пациентов, достигших абсолютного значения SALT ≤20 баллов на 4-й неделе
* Доля пациентов, достигших абсолютной оценки по шкале SALT ≤10 баллов на 24-й неделе

Кроме того, непрерывная вторичная конечная точка SALT - относительное изменение
показателя SALT по сравнению с исходным уровнем на неделях 4, 8, 12, 16, 20 и 24.

*Вроде бы это клиническая.*

3) **Глобальное клиническое впечатление об улучшении (CGI-I)**

Оценка CGI-I будет проводиться исследователем на 12, 16, 20 и 24-й неделях. Текущее состояние пациента оценивается по сравнению с состоянием алопеции до начала лечения на исходном уровне в соответствии с предполагаемыми изменениями. Исследователь может выбрать один из семи вариантов числовых значений: от "Очень сильно ухудшилось" до "Очень сильно улучшилось".

*На мой взгляд, суррогатная точка, пусть и оценивается исследователем. Т.е. вроде бы это является желаемым результатом лечения. Но критерии субъективны.*

4) **Глобальное клиническое впечатление о тяжести состояния (CGI-S)**

Оценка CGI-S будет проводиться исследователем с учетом степени тяжести алопеции
пациента на момент оценки. Исследователь может выбрать один из семи вариантов
числовых значений: от "Сильнейшее выпадение волос" до "Нормальное, без выпадения
волос".

*На мой взгляд, суррогатная точка, пусть и оценивается исследователем*

5) **Глобальное впечатление пациента об улучшении (PGI-I)**

Показатели PGI-I будут оцениваться пациентом на 12, 16, 20 и 24-й неделях. По
сравнению с состоянием алопеции до начала лечения на исходном уровне, текущее
состояние алопеции пациента будет оцениваться в соответствии с его/ее
предполагаемыми изменениями. Пациент может выбрать один из семи числовых
вариантов: от "Очень сильно ухудшилось" до "Очень сильно улучшилось".

*Суррогатная точка*

6) **Глобальное впечатление пациента о тяжести состояния (PGI-S)**

При оценке PGI-S пациент учитывает степень тяжести алопеции (alopecia areata) на
момент оценки. Пациент может выбрать один из семи вариантов числовых значений: от
"Сильнейшее выпадение волос" до "Нормальное, без выпадения волос".

*Суррогатная точка*

7) **BETA** - это шкала, оценивающая общее количество волос в бровях. Шкала используется для каждой брови, где индивидуальный балл брови определяется путем оценки плотности (D) x площади поверхности (SA). Экспертная группа будет проводить центральное считывание данных на основе фотографий бровей и выставлять оценку для каждого пациента с поражением бровей.

Непрерывная вторичная конечная точка - изменение по сравнению с исходным уровнем
показателя BETA на 12-й и 24-й неделях

*Пожалуй суррогатная точка. Хотя тут есть, что измеряем. И вроде бы это является индикатором эффективности лечения. Но насколько я понимаю алопеция со средней и тяжёлой степенью - это не совсем только про брови. А потому всё так это не является конечной целью лечения.*

8) **BELA** - это шкала, оценивающая общее количество волос на ресницах. Шкала
используется для каждого глаза, где общий балл ресниц определяется путем оценки
густоты и распределения на правом и левом глазу. Экспертная группа будет проводить центральное считывание по фотографиям ресниц и выставлять оценку для каждого пациента с поражением ресниц. Исследование BELA будет проводиться на исходном уровне, на 12-й и 24-й неделях.

Непрерывная вторичная конечная точка - изменение по сравнению с исходным уровнем
показателя BELA на 12-й и 24-й неделях

*Суррогатная точка с аналогичным обоснованием как и в пункте выше*

9)  **Качество волос по отзывам пациентов (Quality of Hair Patient Reported
Outcome, QPRO)**

Опросник QPRO представляет собой оценку из четырех пунктов, которая дает
дополнительную информацию о ключевых характеристиках волос пациента и помогает
определить контекст ответа SPRO. Как и в SPRO, ответы на вопросы QPRO оцениваются по шкале 1-5, где 1 = очень доволен, 2 = доволен, 3 = ни доволен, ни недоволен, 4 = недоволен и 5 = очень недоволен.

*Суррогатная точка. Не является конечной целью исследования*

10) **Госпитальная шкала тревоги и депрессии (HADS)**

Опросник HADS состоит из 14 пунктов и предназначен для оценки симптомов тревоги и депрессии у медицинских пациентов. Пункты оцениваются по 4-балльной шкале тяжести. HADS состоит из двух шкал: тревоги (HADS-A) и депрессии (HADS-D).

*Суррогатная точка. Не является конечной целью исследования*
